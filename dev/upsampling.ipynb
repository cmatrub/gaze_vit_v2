# Upsampling: 2 methods
- Method 1: upsample each head, average the cross entropies from each head
- Method 2: mean pool attention logits from each head, upsample, calculate cross entropy

## Method 1

input: 
    gaze_preds: non-softmaxed attn logits, shape: (b h l, l)  
    gaze_targs: shape: (b, N)  
    where b = batch_size, h = number of heads, p = number of patches, l = sqrt(p), N = (og input image size)^2  

i need to get both to (b h N)

```python
import torch
import math
from einops import repeat
```

```python
gaze_preds = torch.tensor([[[1, 2], [3, 4]]])
b, h, p = gaze_preds.shape
l = math.sqrt(p)
patch_size_squared = 4
gaze_preds
```

```python
# upsample
gaze_preds = repeat(gaze_preds, 'b h p -> b h (p patch_square)', patch_square=patch_size_squared) # p * patch_size_squared = N
gaze_preds
```
